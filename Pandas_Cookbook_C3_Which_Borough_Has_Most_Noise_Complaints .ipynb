# 3. Which Borough Has Most Noise Complaints

This is a subset of the of 311 service requests from NYC Open Data.

https://nycopendata.socrata.com/Social-Services/311-Service-Requests-from-2010-to-Present/erm2-nwe9

```python
#---Step 0: Import necessary libraries and render plots inline---

import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline

plt.rcParams['figure.figsize'] = (15, 5)
```

```python
#---Step 1: Read in CSV file---

complaints = pd.read_csv("/Users/cheon/pandas-cookbook-practice/data/311-service-requests.csv")

complaints
```

```python
#---Step 2: Select rows where complaint type is "Noise - Street/Sidewalk"---

noise_complaints = complaints.loc[complaints["Complaint Type"] == "Noise - Street/Sidewalk",:]

noise_complaints
```

```python
#---Step 3: Select rows where complaint type is "Noise - Street/Sidewalk" and borough is "BROOKLYN"---

noise_bool = complaints["Complaint Type"] == "Noise - Street/Sidewalk"
brooklyn_bool = complaints["Borough"] == "BROOKLYN"

complaints[noise_bool & brooklyn_bool].head()
```

```python
#---Step 4: Same condition as above, but select only a few columns---

complaints.loc[noise_bool & brooklyn_bool, ["Complaint Type", "Borough", "Created Date", "Descriptor"]].head(10)
```

```python
#---Step 5: Determine which borough has the most noise complaints---

print(complaints.loc[noise_bool,"Borough"].value_counts())

print(complaints.loc[noise_bool,"Borough"].value_counts(normalize=True))
```

```python
#---Step 6: Determine which borough has greatest proportion of most noise complaints relative to all complaints for that borough---

noise_complaint_counts = complaints.loc[noise_bool, "Borough"].value_counts()
complaint_counts = complaints.loc[:,"Borough"].value_counts()

noise_complaint_prop_borough = noise_complaint_counts / complaint_counts

noise_complaint_prop_borough
```

```python
#---Step 7: Plot proportion of noise complaints to all complaints for each borough---

noise_complaint_prop_borough.plot(kind="bar")
```

Manhattan really does complain more about noise than the other boroughs.
