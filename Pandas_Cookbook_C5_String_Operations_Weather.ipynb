# 5. String Operations with Weather Data

By the end of this chapter, we're going to have downloaded all of Canada's weather data for 2012, and saved it to a CSV.

We'll do this by downloading it one month at a time, and then combining all the months together.

Here's the temperature every hour for 2012!

```python
#---Step 0: Import necessary libraries and render plots inline---

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline

plt.rcParams['figure.figsize'] = (15, 3)
plt.rcParams['font.family'] = 'sans-serif'
```

```python
#---Step 1: Read in CSV file with appropriate parsing arguments---

# Parse dates in the "Date" column
# Set index to be the "Date/Time" column

weather_2012_final = pd.read_csv("/Users/cheon/pandas-cookbook-practice/data/weather_2012.csv", parse_dates=["Date/Time"], index_col="Date/Time")

weather_2012_final['Temp (C)'].plot(figsize=(15, 6))
```

```python
#---Step 2: Plot temperature by hour of day---

temperatures = pd.DataFrame(weather_2012_final["Temp (C)"])
temperatures["Hour"] = temperatures.index.hour

temperatures.groupby("Hour").aggregate("median").plot()
```
